### *Determinado* os Outliers** (REVER)**

```python
 #interquartile range will be used
imdb_q1 = ds["IMDB Score"].quantile(0.25) #first quartile
imdb_q3 = ds["IMDB Score"].quantile(0.75) #third quartile

imdb_iqr = imdb_q3-imdb_q1

imdb_up = imdb_q3 + 1.5 * imdb_iqr #upper limit
imdb_low = imdb_q1 - 1.5 * imdb_iqr #lower limit

outlier_imdb = ds[(ds["IMDB Score"] < imdb_low) | (ds["IMDB Score"] > imdb_up)]
count_imdb = outlier_imdb.count()["IMDB Score"]

print(f"imdb_outlier_count: {count_imdb}")

sns.boxplot(x = ds['IMDB Score'])
plt.show()
```
